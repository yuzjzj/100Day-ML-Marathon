# Rectified Linear Unit- Relu 

f(x)=max(0,x)


```python
import numpy as np
from numpy import *
import matplotlib.pylab as plt
%matplotlib inline

def ReLU(x):
    return max(0,x)
def dReLU(x):
    if(x>0):
        return 1
    return 0
```

```python
a=-2
b=2
N=100
x = np.linspace(a,b,N+1)
y = list(map(ReLU,x))
dy= list(map(dReLU,x))
```

```python
plt.plot(x,y)
plt.plot(x,dy,"--")
plt.show()
```
